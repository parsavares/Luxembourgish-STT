## II. Record Input Sample

```python
from IPython.display import Javascript
from google.colab import output
from base64 import b64decode

# JavaScript to create audio recorder and record a clip
js = Javascript('''
async function recordAudio(duration) {
  const stream = await navigator.mediaDevices.getUserMedia({audio: true});
  const mediaRecorder = new MediaRecorder(stream);
  let audioChunks = [];

  mediaRecorder.addEventListener("dataavailable", event => {
    audioChunks.push(event.data);
  });

  setTimeout(() => mediaRecorder.stop(), duration * 1000); // Stop recording after duration
  mediaRecorder.start();

  return new Promise(resolve => {
    mediaRecorder.addEventListener("stop", () => {
      const audioBlob = new Blob(audioChunks);
      const audioUrl = URL.createObjectURL(audioBlob);
      const reader = new FileReader();
      reader.readAsDataURL(audioBlob);
      reader.onloadend = () => {
        const base64data = reader.result;
        resolve(base64data);
      };
    });
  });
}
''')

def record_audio(duration=5):
    display(js)  # Display the JavaScript interface for recording
    print('Speak now...')
    audio_data = output.eval_js('recordAudio({})'.format(duration))
    binary = b64decode(audio_data.split(',')[1])

    # Save the audio data to a file
    with open('audio_recorded.wav', 'wb') as f:
        f.write(binary)
    print('Audio recording saved as "audio_recorded.wav"')

# Use this to record 5 seconds of audio
record_audio(10)
```

```python
import torchaudio

def preprocess_audio(audio_path):
    waveform, sample_rate = torchaudio.load(audio_path)
    # Check if the audio is stereo and convert it to mono if necessary
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)
    # Resample to the sample rate expected by the model (e.g., 16kHz)
    resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
    waveform = resampler(waveform)
    return waveform
```

## 0.II. Prediction-function implementation

```python
#for first model wav2vec2-large-xlsr-53-842h-luxembourgish-14h
import time
def predict(transcript_df):
    predictions = []
    for _, row in transcript_df.iterrows():
        file_path = f'/content/drive/MyDrive/work/STT/meryem/kapitel_1/{row["filename"]}'
        print(f"Processing file: {file_path}")  # Debugging output
        waveform = preprocess_audio(file_path)
        inputs = processor(waveform.squeeze(), sampling_rate=16_000, return_tensors="pt").input_values
        with torch.no_grad():
            tic = time.perf_counter()
            logits = model(inputs).logits
        predicted_ids = torch.argmax(logits, dim=-1)
        transcription = processor.decode(predicted_ids[0])
        toc = time.perf_counter()
        print(f"timing:  {toc - tic:3f} seconds")
        predictions.append(transcription)
    return predictions
```

```python
#for the second model whisper_large_lb_ZLS_v4_38h
def predict2(transcript_df):
    predictions = []
    for _, row in transcript_df.iterrows():
        file_path = f'/content/drive/MyDrive/work/STT/meryem/kapitel_1/{row["filename"]}'
        print(f"Processing file: {file_path}")
        waveform = preprocess_audio(file_path)
        processed = processor2(waveform.squeeze(), sampling_rate=16000, return_tensors="pt")
        input_features = processed['input_features']

        with torch.no_grad():
            tic = time.perf_counter()
            try:
                # Directly use generate without specifying decoder_input_ids
                generated_ids = model2.generate(input_features)
                transcription = processor2.batch_decode(generated_ids)[0]
            except Exception as e:
                print("Error during generation:", e)
                continue  # Skip if error in generation

            toc = time.perf_counter()
            print(f"Timing: {toc - tic:.3f} seconds")
            predictions.append(transcription)
    return predictions
```

```python
import os
import torch
import torchaudio
import pandas as pd
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq
from google.colab import drive
from jiwer import wer, Compose, ToLowerCase, RemoveMultipleSpaces, RemovePunctuation, Strip
import time

# Token
os.environ['HF_TOKEN'] = "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"  # UNILU token

# Load processor and model, move model to GPU
processor2 = AutoProcessor.from_pretrained("ZLSCompLing/whisper_large_lb_ZLS_v4_38h")
model2 = AutoModelForSpeechSeq2Seq.from_pretrained("ZLSCompLing/whisper_large_lb_ZLS_v4_38h").to('cuda')
```

```python
# Mount Google Drive
drive.mount('/content/drive')

# Load transcript data
transcript_path = '/content/drive/MyDrive/work/STT/meryem/kapitel_1/chapter_1_transcriptions.txt'
transcripts = pd.read_csv(transcript_path, sep='\t', header=None, names=['filename', 'transcription'])
transcripts['filename'] = transcripts['filename'].str.strip()  # Remove any leading/trailing spaces
```

```python
def preprocess_audio(audio_path):
    waveform, sample_rate = torchaudio.load(audio_path)
    # Check if the audio is stereo and convert it to mono if necessary
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)
    # Resample to the sample rate expected by the model (e.g., 16kHz)
    resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
    waveform = resampler(waveform)
    return waveform

def predict2(transcript_df):
    predictions = []
    for _, row in transcript_df.iterrows():
        file_path = f'/content/drive/MyDrive/work/STT/meryem/kapitel_1/{row["filename"]}'
        print(f"Processing file: {file_path}")
        waveform = preprocess_audio(file_path).to('cuda')  # Move waveform to GPU
        processed = processor2(waveform.squeeze().cpu(), sampling_rate=16000, return_tensors="pt")
        input_features = processed['input_features'].to('cuda')  # Ensure input features are on GPU

        with torch.no_grad():
            tic = time.perf_counter()
            try:
                # Directly use generate without specifying decoder_input_ids
                generated_ids = model2.generate(input_features)
                transcription = processor2.batch_decode(generated_ids.cpu())[0]  # Move generated_ids to CPU for decoding
            except Exception as e:
                print("Error during generation:", e)
                continue  # Skip if error in generation

            toc = time.perf_counter()
            print(f"Timing: {toc - tic:.3f} seconds")
            predictions.append(transcription)
    return predictions

```

```python
from transformers import AutoProcessor, AutoModelForCTC, TrainingArguments, Trainer
import torch
from datasets import load_dataset, Audio
import torch.nn.functional as F
from dataclasses import dataclass
from typing import Any, Dict, List, Union

# Ensure the 'audio' column is correctly formatted
dataset = dataset.cast_column("audio", Audio())

# Preprocess the dataset
def prepare_dataset(batch):
    audio = batch["audio"]
    batch["input_values"] = processor(audio["array"], sampling_rate=audio["sampling_rate"]).input_values[0]
    batch["labels"] = processor.tokenizer(batch["transcription"]).input_ids
    return batch

dataset = dataset.map(prepare_dataset, remove_columns=dataset["train"].column_names)

# Custom data collator
@dataclass
class DataCollatorCTCWithPadding:
    processor: AutoProcessor

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        input_features = [{"input_values": feature["input_values"]} for feature in features]
        label_features = [{"input_ids": feature["labels"]} for feature in features]

        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")
        with self.processor.as_target_processor():
            labels_batch = self.processor.pad(label_features, return_tensors="pt")

        # Replace padding with -100 to ignore loss correctly
        labels = labels_batch["input_ids"].masked_fill(labels_batch.attention_mask.ne(1), -100)

        batch["labels"] = labels

        return batch

data_collator = DataCollatorCTCWithPadding(processor=processor)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    learning_rate=1e-4,
    weight_decay=0.01,
    save_total_limit=2,
    fp16=True,  # Use mixed precision for faster training
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["validation"],
    data_collator=data_collator,
    tokenizer=processor.feature_extractor,
)

# Move model to GPU
if torch.cuda.is_available():
    model.to("cuda")

# Start training
trainer.train()
```

```python
# Define a function to transcribe audio chunk to text
def transcribe_audio_chunk(chunk, sample_rate=16000):
    inputs = processor2(chunk, sampling_rate=sample_rate, return_tensors="pt")
    input_features = inputs.input_features.to(device)
    start_token_id = model2.config.decoder_start_token_id
    decoder_input_ids = torch.tensor([[start_token_id]], device=device)

    with torch.no_grad():
        outputs = model2.generate(input_features=input_features, decoder_input_ids=decoder_input_ids, max_length=448)

    transcription = processor2.batch_decode(outputs, skip_special_tokens=True)[0]

    return transcription

# Define a function to get response from ChatGPT
def get_chatgpt_response(transcription):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=transcription,
        max_tokens=150
    )
    return response.choices[0].text.strip()

# Function to list available input devices
def list_input_devices():
    p = pyaudio.PyAudio()
    info = p.get_host_api_info_by_index(0)
    num_devices = info.get('deviceCount')
    for i in range(0, num_devices):
        device_info = p.get_device_info_by_host_api_device_index(0, i)
        if device_info.get('maxInputChannels') > 0:
            print(f"Input Device ID {i} - {device_info.get('name')}")
    p.terminate()

# List available input devices
list_input_devices()

# Real-time speech recognition function
def live_speech_recognition(input_device_index):
    CHUNK = 1024  # Buffer size
    FORMAT = pyaudio.paInt16  # 16-bit resolution
    CHANNELS = 1  # Mono channel
    RATE = 16000  # 16kHz sample rate
    RECORD_SECONDS = 5  # Recording duration

    p = pyaudio.PyAudio()

    stream = p.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True,
                    input_device_index=input_device_index,
                    frames_per_buffer=CHUNK)

    print("Listening...")

    try:
        while True:
            frames = []

            for _ in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
                data = stream.read(CHUNK)
                frames.append(data)

            audio_data = np.frombuffer(b''.join(frames), dtype=np.int16)
            audio_data = audio_data / np.iinfo(np.int16).max

            transcription = transcribe_audio_chunk(audio_data)
            response = get_chatgpt_response(transcription)

            print("You:", transcription)
            print("ChatGPT:", response)

    except KeyboardInterrupt:
        print("Stopped listening.")

    stream.stop_stream()
    stream.close()
    p.terminate()

# Specify the correct input device index (replace with your chosen device ID)
input_device_index = 1  # Replace with your chosen device ID from the list
```

```python
# Start real-time speech recognition in a separate thread
thread = threading.Thread(target=live_speech_recognition, args=(input_device_index,))
thread.start()
```
